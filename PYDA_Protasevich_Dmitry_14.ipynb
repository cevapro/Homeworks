```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
"""В качестве датасета использован "World Happiness Report" за 2019 год"""
happy_index = pd.read_csv('C:/IT/DATA_for_homeworks/014_Data-visualization/2019.csv')
happy_index.head()
```

```python
plt.plot(happy_index['Score'], happy_index['Perceptions of corruption'])
plt.title('Зависимость уровня коррупции и рейтинга страны')
plt.xlabel('Score')
plt.ylabel('Perceptions of corruption')
```

Гипотеза: чем выше уровень счастья (рейтинг страны), тем ниже уровень коррупции в стране (выше показатель Perceptions of corruption). 
Возможно, верно и обратное.
На графике видим проявление зависимости показателей только при рейтинге страны выше 6,5 баллов.
Чтобы проверить это построим график в виде точечной диаграммы.

```python
happy_index.plot(kind='scatter', 
        x='Score', 
        y='Perceptions of corruption', 
        s=happy_index.Score*2,
        title='Зависимость уровня коррупции и рейтинга страны',
       )
```

Точечная диаграмма несколько корректирует представление. Чем выше рейтинг страны тем больше вероятность снижения уровня коррупции, но это не является гарантией. Однако в целом при рейтинге страны выше 6,5 сильно растет вероятность меньшей коррупции. Интересно, что при приближении рейтинга от 3 к 6, общее значение коррупции увеличивается.

```python
happy_index.plot(kind='scatter', 
        x='Score', 
        y='Perceptions of corruption', 
        s=happy_index.Score*2,
        title='Количество стран и уровень коррупции',
       )
plt.plot(happy_index.Score, happy_index.index*0.005)
```

Количество стран имеющих высокий рейтинг резко снижается к значению рейтинга 6,5. Далее количество стран уменьшается не с таким большим темпом. Гипотеза: страны с рейтингом выше 6,5 принципиально отличаются от стран с рейтингом ниже и по другим показателям жизни (состояние медицины, продолжительност жизни и пр.).

```python
happy_index.Score.plot(kind='hist', 
                     bins=10,
                     xticks=range(1, 8))
```

На гистограмме видим распределение количества стран занимающих тот или иной рейтинг. Большая часть стран находится в рейтинге от 4.5 до 6. Гипотеза: уровень жизни в странах с рейтингом ниже 4.5 сильно ниже, а стран с рейтингом выше 6 сильно выше, чем в странах со средним рейтингом.
