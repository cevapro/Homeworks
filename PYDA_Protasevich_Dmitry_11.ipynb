```python
import pandas as pd

rating_data = pd.read_csv('C:/IT/DATA_for_homeworks/Pandas_library/ml-25m/ratings.csv')
rating_data.head()
```

```python
def class_c(x):
    """оценка рейтинга"""
    if x >= 4.5:
        param = 'высокий рейтинг'
    elif x <= 2:
        param = 'низкий рейтинг'
    else:
        param = 'средний рейтинг'
    return param

rating_data['class'] = rating_data['rating'].apply(class_c)
rating_data
```

# Задание 2

```python
keywords_data = pd.read_csv('C:/IT/DATA_for_homeworks/Pandas_functions_groupby/keywords.csv')
keywords_data.head()
```

```python
def region(geo):
    """определение региона"""
    if geo in ['москва', 'тула', 'ярославль']:
        geo_data = 'Центр'
    elif geo in ['петербург', 'псков', 'мурманск']:
        geo_data = 'Северо-Запад'
    elif geo in ['владивосток', 'сахалин', 'хабаровск']:
        geo_data = 'Дальний Восток'
    else:
        geo_data = 'undefined'
    return geo_data

keywords_data['region'] = keywords_data['keyword'].apply(region)
print(keywords_data.head())
keywords_data.region.unique()
```
